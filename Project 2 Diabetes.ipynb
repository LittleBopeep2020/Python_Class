```python
#Gathering data for All ages population taking diabetes medication. 
Race_Meds = df[(df['race'].isin(['AfricanAmerican','Asian','Caucasian','Hispanic','Other'])&
                   (df['age']) &
                   (df['diabetesMed']=='Yes'))]
```

```python
#Verifying data
Race_Meds.head(10)
```

```python
#Data Gathering
data_Readmission = Race_Meds[['race','gender','age','diabetesMed','readmitted']]
data_Readmission.groupby(['race'])['readmitted'].count().sort_values(ascending=True).plot(kind='bar')
```

```python
#Collecting data count of population by race
Pop_by_race = data_Readmission.groupby(['race']).count()
Pop_by_race
```

```python
#Collecting data count of readmission by race
No_readmitted = data_Readmission[data_Readmission['readmitted'] == 'Yes'].groupby('race').count()
No_readmitted
```

```python
#Trying to get the percentage of readmitted vs. population by race
Percent_readmitted = No_readmitted/Pop_by_race
Percent_readmitted
```

```python

Percent_readmitted.describe()
```

```python
#Illustration of data
Percent_readmitted['readmitted'].sort_values(ascending=False).plot(kind='bar', color=['red','green','yellow','blue','purple'])
plt.xticks(rotation=45)
plt.title('Readmission by Race')
plt.xlabel('Race and Ethnicity')
plt.ylabel('Percentage of readmission')
```

```python
#In conclusion,the data shows that Caucasians had the highest rate of readmission 
#  with the 48% of the total population of taking prescribed diabetes medication.
#  However, 36% of total population of Asians that are taking prescribed diabetes  
#  medication and had the lowest rate.  These differences may be influenced by a  
#  combination of genetic, socioeconomic, and healthcare access factors.
```
